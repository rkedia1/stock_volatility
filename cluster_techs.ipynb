```python
from analysis import ClusterFinancials
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

cf = ClusterFinancials()
```

```python
ct = cf.kmeans
x, y = ct.inertia()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Sum of Squared Within-Cluster Distances")
plt.title("Clustering Quality based on Inertia")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
x, y = ct.calinski_harabasz()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Calinski-Harabssz Score")
plt.title("Clustering Quality based on Calinski-Harabsz Scores")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
x, y = ct.silhouette()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Clustering Quality based on Silhouette Scores")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
x, y = ct.davis_bouldin()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Davis-Bouldin Score")
plt.title("Clustering Quality based on Davis-Bouldin Scores")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
assignment, km = ct.cluster(n_clusters=5)
labels = {0: [], 1: [], 2: [], 3: [], 4:[]}
for i in range(len(assignment)):
    labels[km.labels_[i]].append(assignment[i])
print(labels)
```

```python
gmm = list(cf.gmm.cluster())
cluster_count = max(set([i[1] for i in gmm]))
gmm_dict = dict()
for i in range(0, cluster_count+1):
    gmm_dict[i] = []
for i in gmm:
    gmm_dict[i[1]].append(i[0])
print(gmm_dict)
```

### Visualizing GMM clustering using PCA

```python
results = cf.pca.reduce()
df = pd.DataFrame(columns=["PC1", "PC2", "Cluster"])
df["PC1"] = results[:, 0]
df["PC2"] = results[:, 1]
df["Cluster"] = [f"Cluster {i[1]}" for i in gmm]
plt.figure(figsize=(5, 5))
sns.scatterplot(
    x="PC1",
    y="PC2",
    hue="Cluster",
    palette=sns.color_palette("hls", cluster_count+1),
    data=df.sort_values(by="Cluster"),
    legend="full",
)
plt.show()
```

### Visualizing GMM clustering using t-SNE

```python
results = cf.tsne.reduce()
df = pd.DataFrame(columns=["tsne-1", "tsne-2", "Cluster"])
df["tsne-1"] = results[:, 0]
df["tsne-2"] = results[:, 1]
df["Cluster"] = [f"Cluster {i[1]}" for i in gmm]
plt.figure(figsize=(5, 5))
sns.scatterplot(
    x="tsne-1",
    y="tsne-2",
    hue="Cluster",
    palette=sns.color_palette("hls", cluster_count+1),
    data=df.sort_values(by="Cluster"),
    legend="full",
)
plt.show()
```

```python
import black
import jupyter_black

jupyter_black.load(lab=False)

from analysis import ClusterFinancials
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

cf = ClusterFinancials()
```

```python
ct = cf.kmeans
x, y = ct.inertia()
plt.plot(x, y, "bx-")
plt.xlabel("Values of K")
plt.ylabel("Inertia")
plt.title("The Elbow Method using Inertia")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
x, y = ct.calinski_harabasz()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Calinski-Harabssz Score")
plt.title("Clustering Quality based on Calinski-Harabsz Scores")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
x, y = ct.silhouette()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Clustering Quality based on Silhouette Scores")
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
x, y = ct.davis_bouldin()
plt.plot(x, y, "bx-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Davis-Bouldin Score")
plt.title("Clustering Quality based on Davis-Bouldin Scores")
plt.xticks(ticks=x, labels=x)
plt.show()
```

### Visualizing GMM clustering using t-SNE

```python
results = cf.tsne.cluster()
```

```python
df = pd.DataFrame(columns=["TC1", "TC2", "Cluster"])
df["TC1"] = results[:, 0]
df["TC2"] = results[:, 1]
df["Cluster"] = [f"Cluster {i[1]}" for i in gmm]
plt.figure(figsize=(5, 5))
sns.scatterplot(
    x="TC1",
    y="TC2",
    hue="Cluster",
    palette=sns.color_palette("hls", 8),
    data=df.sort_values(by="Cluster"),
    legend="full",
)
plt.show()
```

```python
from analysis import KmeansTechnicals, GMMTechnicals, TSNETechnicals
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
ct = KmeansTechnicals()
x, y = ct.inertia()
plt.plot(x, y, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
ct = KmeansTechnicals()
x, y = ct.calinski_harabasz()
plt.plot(x, y, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Calinski-Harabssz Score')
plt.title('Clustering Quality based on Calinski-Harabsz Scores')
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
ct = KmeansTechnicals()
x, y = ct.silhouette()
plt.plot(x, y, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Clustering Quality based on Silhouette Scores')
plt.xticks(ticks=x, labels=x)
plt.show()
```

```python
ct = KmeansTechnicals()
x, y = ct.davis_bouldin()
plt.plot(x, y, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Davis-Bouldin Score')
plt.title('Clustering Quality based on Davis-Bouldin Scores')
plt.xticks(ticks=x, labels=x)
plt.show()
```
